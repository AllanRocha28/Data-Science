# Panorama do COVID-19 no Brasil e no mundo






Em dezembro de 2019, a **Organização Mundial da Saúde** (**OMS**) foi alertada sobre um possível novo vírus descoberto a partir de casos de pneumonia na cidade de Wuhan, na República Popular da China. A nova cepa (tipo) do coronavírus foi identificada e nomeada de **SARS-CoV-2** e causa a doença **COVID-19**.

Em janeiro de 2020 a **OMS** a declarou como sendo um problema internacional e em março do mesmo ano a caracterizou como pandemia.

A **COVID-19** é uma doença que pode ser transmitida a outras pessoas através de **espirros**, **tosses** ou contato com **superfícies de objetos contaminados**.

A doença pode passar despercebida em muitos casos por não apresentar sintomas, **casos assintomáticos**, e em muitos casos em que apresentam indícios podem ser confundidos com resfriado comum, por apresentarem os mesmos sintomas. Porém o quadro pode evoluir rapidamente para um caso mais drástico aonde o infectado pode apresentar **falta de paladar**, **problemas neurológicos**, **grave dificuldade para respirar**, a ponto de precisar de respiração mecânica no hospital, podendo chegar a **óbito**.

A partir das declarações da **OMS** buscou-se diminuir a propagação do vírus e iniciou uma corrida contra o tempo para compreender e buscar vacinas que pudessem combater a **COVID-19**.

No Brasil o primeiro caso detectado foi de um homem de 61 anos que esteve na Itália em fevereiro de 2020; em março foram registrados o primeiro caso de transmissão interna no país e a primeira morte causada pelo vírus.

### **Quais são as principais variáveis relacionadas com o aumento no número de mortes por Covid-19?**

Responderemos a essa pergunta selecionando algumas *features*, em um novo dataframe, que possuem alguma relação com a variável alvo (*`new_deaths`*),segundo os profissionais da área da saúde.

Pode ser visto no gráfico de *heatmap* que com o aumento no número de novas mortes (*`new_deaths`*) há um aumento na resposta dos governos em relação a medidas de isolamento e fechamentos de instituições (*`stringency_index`*).

O número de mortes também tem uma correlação positiva com a idade da pessoa, o que explica o número expressivo de mortes em países que tem uma população mais envelhecida.

```
# Seleção das variáveis utilizadas para o gráfico de correlação

df_heatmap = df.groupby(['date'], as_index=False)[['reproduction_rate', 'total_deaths', 'new_cases', 'new_deaths', 'positive_rate',
                                                  'new_tests', 'total_tests', 'people_vaccinated', 'people_fully_vaccinated', 'total_vaccinations',
                                                  'stringency_index', 'extreme_poverty', 'aged_65_older', 'aged_70_older', 'female_smokers', 'male_smokers']].sum()

# Criando a tabela de correlação entre as variáveis
df_heatmap_corr = df_heatmap.corr()

# Número de variáveis com maior nível de correlação
k = 6

# Buscando as 6 variáveis com maior correlação com 'new_deaths'
cols = df_heatmap_corr.nlargest(k, 'new_deaths')['new_deaths'].index
cm = np.corrcoef(df_heatmap[cols].values.T)

# plotando o heatmap
f, ax = plt.subplots(figsize = (14,8))
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)

# Rodapé
plt.text(-0.6, -0.40, 'Fonte dos dados: https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv\n'
                  'Autor: Allan Amaral da Hora Rocha',
         fontsize=12, color='#B5B5B5', weight='bold', transform=ax.transAxes)

# salvar imagem
fig = plt.gcf()
fig.tight_layout() # ajusta o tamanho da figura
plt.show()  # plotar gráfico
fig.savefig('gráfico1_projeto_covid.jpg', format='jpg')  #Salvando em pdf
```

### **Análise da doença nos continentes**

```
# configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'

# definindo paleta de cores para os gráficos
LARANJA1, LARANJA2, LARANJA3, LARANJA4, LARANJA5 = "#ED9D08", "#F68A06", "#F4730B", "F96209", "#F94001"
CINZA1, CINZA2, CINZA3 = "#3857FA", "#A2A1A3", "#787878"
AZUL1, AZUL2 = "#8db7eb", "#253760"
VERMELHO1 = "#A8240F"

# definindo paleta de cores para as bordas dos gráficos
EDGE_1 = [CINZA3, VERMELHO1, CINZA3, CINZA3, CINZA3]
EDGE_2 = [AZUL2, VERMELHO1, AZUL2, AZUL2, AZUL2]

# definindo conjunto de paleta de cores para os gráficos
CORES1 = [CINZA2, CINZA3, CINZA2, CINZA2, CINZA2]
CORES2 = [AZUL1, AZUL2, AZUL1, AZUL1, AZUL1]
```

Algumas *features* apresentam valores ausentes (NaN) e que precisam ser alterados pra podermos começar a análise desse estudo.

Iremos substituir esses valores (NaN) por 0.

```
# Substituindo os valores ausentes (NaN) por 0
df2_continent_cases = df2.fillna(0)

# Seleção das features e soma dos valores para cada continente
df2_continent_cases = df2.groupby(['continent'], as_index=False)[['total_cases', 'total_deaths', 'people_vaccinated', 'people_fully_vaccinated', 'total_vaccinations', 'total_boosters']].sum()

df2_continent_cases
```

Vemos que o continente americano está dividido em *`North America`* e *`South America`*. Vamos transformar esses dois valores em um único chamado America. 

```
# Transformando as variáveis South America e North America em America na feature continent
df2_continent_cases.replace(['South America', 'North America'], value='America', inplace=True)
```

```
# Passando a feature continent como index para juntas as variáveis America
df2_continent_cases = df2_continent_cases.groupby(['continent'], as_index=True)[['total_cases', 'total_deaths', 'people_vaccinated', 'people_fully_vaccinated', 'total_vaccinations', 'total_boosters']].sum()

df2_continent_cases
```

```
# Criando grupos pelos continentes com porcentagens para cada variável 
porcentagem_1 = (df2_continent_cases['total_cases']/df2_continent_cases['total_cases'].sum())*100

porcentagem_2 = (df2_continent_cases['total_deaths']/df2_continent_cases['total_deaths'].sum())*100

porcentagem_3 = (df2_continent_cases['people_vaccinated']/df2_continent_cases['people_vaccinated'].sum())*100

porcentagem_4 = (df2_continent_cases['people_fully_vaccinated']/df2_continent_cases['people_fully_vaccinated'].sum())*100

porcentagem_5 = (df2_continent_cases['total_vaccinations']/df2_continent_cases['total_vaccinations'].sum())*100

porcentagem_6 = (df2_continent_cases['total_boosters']/df2_continent_cases['total_boosters'].sum())*100
```

No gráfico seguinte podemos perceber que tanto o número de casos quanto o de morte devido a Covid-19 foram maiores no continente Americano do que nos demais continentes e o número de mortes devido a doença foi maior que os de casos detectados, apreentando um comportamento diferente do ocorrido nos dois outros continentes com maior número de casos da doença.

```
fig, ax = plt.subplots(figsize=(16,8))

x = np.arange(5)
width_bar = 0.45

graph1 = ax.bar(x-0.010-width_bar/2, porcentagem_1, width_bar, label="nº de casos confirmados", color=CINZA3, edgecolor = EDGE_1, linewidth=3.3);
graph2 = ax.bar(x+0.010+width_bar/2, porcentagem_2, width_bar, label="nº de mortes devido a Covid", color=AZUL2, edgecolor = EDGE_2, linewidth=3.3);

# labels do eixo x
ax.set_xticks(x)
ax.set_xticklabels(['Africa', 'America', 'Asia', 'Europe', 'Oceania'], fontsize=12)

# remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# removendo o eixo y
plt.gca().axes.get_yaxis().set_visible(False)

# remover xticks do eixo x
plt.tick_params(
    axis='both',         
    which='both',     
    bottom=False,      
    top=False,         
    labelbottom=True)

#criar função para dar labels automaticamente
def autolabel(graphs):
    for graph in graphs:
        height = graph.get_height()
        ax.annotate(
                    '{:0.1f}%'.format(height),
                    xy=(graph.get_x() + graph.get_width()/2,  height-0.3),
                    xytext=(0, -20),  # offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=13,
                    color="white",
                    )
        
# plotar labels dentro das barras
autolabel(graph1)
autolabel(graph2)

#Título
plt.text(-1, 50, 'Casos de Covid no mundo',
         fontsize=25, color=AZUL2, weight='bold')

# plotar subtitulo
plt.text(-1, 49, 'Gráfico do número de casos confirmados e\n'
         'de morte nos continentes devido a Covid.',
         verticalalignment="top",
         fontsize=12,
         color=CINZA3)

# anotações
plt.text(1.5, 45.5,
         "$\\bf{AVANÇO DA DOENÇA}$\n"
         "o número de mortes no continente americano\n"
         "teve um avanço mais rápido do que a detecção\n"
         "dos casos."
         ,
         verticalalignment="top",
         fontsize=13,
         color=AZUL2)

ax.set_facecolor("white")
#legenda
ax.legend(fontsize=16, loc='best')

# Rodapé
plt.text(0, -0.16, 'Fonte dos dados: https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv\n'
                  'Autor: Allan Amaral da Hora Rocha',
         fontsize=12, color='#B5B5B5', weight='bold', transform=ax.transAxes)

# salvar imagem
fig = plt.gcf()
fig.tight_layout() # ajusta o tamanho da figura
plt.show()  # plotar gráfico
fig.savefig('gráfico2_projeto_covid.jpg', format='jpg')  #Salvando em pdf
```

O gráfico abaixo mostra a quantidade de vacinas aplicadas em cada continente.
